```python
import cv2

file = open("/Users/ali/Downloads/UCF_Crimes-Train-Test-Split-2/Action_Regnition_splits/ClassIDs.txt")
lines = file.read().split("\n")
file.close()
classes = {}
classes_temp = [l.split(" ") for l in lines if l!=""]
for i in classes_temp:
   classes[i[0]] = i[1] 
classes
r = "/Users/ali/Desktop/train_vid/"
```

```python
r = "/Users/ali/Desktop/train_vid/"
file = open(r + "train.txt", "r")
lines = file.read().split("\n")
file.close()
file = open(r + "/train_final.txt", "a")
for l in lines:
    l = l[0:-1]
    row = l
    cap = cv2.VideoCapture(r+l)
    frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    row += " " + str(frames)
    cat = l.split("/", 1)[0]
    label = classes[cat]
    row += " " + str(label) + "\n"
    file.write(row)
    print(row)
file.close()
```

```python
file = open(r + "test.txt", "r")
lines = file.read().split("\n")
file.close()
file.close()
file = open(r + "/test_final.txt", "a")
for l in lines:
    l = l[0:-1]
    row = l
    cap = cv2.VideoCapture(r+l)
    frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    row += " " + str(frames)
    cat = l.split("/", 1)[0]
    label = classes[cat]
    row += " " + str(label) + "\n"
    file.write(row)
    print(row)
file.close()
```

```python
def write_classes(path_from, path_to):
    file = open(r + path_from, "r")
    lines = file.read().split("\n")
    file.close()
    file = open(r + path_to, "a")
    for l in lines:
        l = l.split(" ")[-1] + "\n"
        file.write(l)
    file.close
```

```python
write_classes("/test_final.txt", "y_val.txt")
```

```python
write_classes("/train_final.txt", "y_train.txt")
```
